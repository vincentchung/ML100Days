```python
def img2arr_fromURLs(url_list, resize = False):
    for url in url_list:
        response = requests.get(url)
        print(response.status_code)
        if(response.status_code==200):
            imglst.append(Image.open(BytesIO(response.content)))
        else:
            imglst.append(None)
        #plt.imshow(imglst[i])
    return imglst
```

```python

url="https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt"

urldata=requests.get(url).text
urldata_list=([list(map(str, x.split("\t"))) for x in urldata.split("\n")])
urlDF=pd.DataFrame(urldata_list,columns=["pic","url"])

imglst = img2arr_fromURLs(urlDF.loc[0:4]['url'])
#fetch the pic url part to show the img
for img in imglst:
    if(img is not None):
        plt.imshow(img)



```
